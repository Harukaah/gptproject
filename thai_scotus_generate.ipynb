# Thai Supreme Court Style Generator

```

import torch
from model import GPT, GPTConfig
from utils import get_tokenizer

ckpt_path = "checkpoints/gpt_step200000.pt"
device = "cuda" if torch.cuda.is_available() else "cpu"

ckpt = torch.load(ckpt_path, map_location=device)
cfg = GPTConfig(**ckpt["config"])
model = GPT(cfg).to(device)
model.load_state_dict(ckpt["model"])
model.eval()

tok = get_tokenizer("airesearch/wangchanberta-base-att-spm-uncased")

def generate_text(prompt, max_new_tokens=250, temperature=0.8, top_p=0.9):
    ids = tok.encode(prompt, add_special_tokens=False)
    idx = torch.tensor([ids], device=device)
    with torch.no_grad():
        out = model.generate(idx, max_new_tokens=max_new_tokens, temperature=temperature, top_p=top_p)
    return tok.decode(out[0].tolist(), skip_special_tokens=True)

print(generate_text("จำเลยรับเงินจำนวนดังกล่าวแล้วได้ขี่รถจักรยานออกจากร้านผู้เสียหาย"))
```
